```python
# Q : 대륙별 총 인구수를 내림차순으로 정렬해서 출력
QUERY = """
    SELECT continent, sum(population) AS sum_pop
    FROM country
    GROUP BY continent
    order by sum_pop desc
    limit 5
"""
pd.read_sql(QUERY,db)
```

```python
# Having과 where 차이
QUERY = """
    SELECT continent, sum(population) AS population
    FROM country
    where population >= 50000*10000
    GROUP BY continent
    order by population desc
    limit 5
"""
pd.read_sql(QUERY,db)
```

```python
# Having과 where 차이
QUERY = """
    SELECT continent, sum(population) AS population
    FROM country
    GROUP BY continent
    having population >= 50000*10000
    order by population desc
    limit 5
"""
pd.read_sql(QUERY,db)
```

```python
# Quiz
# 국가코드, 국가이름, 도시이름, 국가인구수, 도시인구수 출력
QUERY = """
    SELECT city.countrycode
           , country.name AS country_name
           , city.name AS city_name
           , country.population AS country_population
           , city.population AS city_population
           , countrylanguage.language
    FROM city
    JOIN country
    ON city.countrycode = country.code
    join countrylanguage
    on country.code = countrylanguage.countrycode
"""
df = pd.read_sql(QUERY, db)
df.head()
```

```python
# UNION : 두개의 쿼리를 실행한 결과를 합쳐서 출력 + 중복 데이터 제거
QUERY = """
    SELECT name
    FROM user
    UNION
    SELECT addr_name
    FROM addr
"""
pd.read_sql(QUERY, db)
```

```python
# OUTER JOIN : left join과 right join을 union하면됨.
QUERY = """
    SELECT user.user_id, user.name, addr.addr_name
    FROM user
    LEFT JOIN addr
    ON user.user_id = addr.user_id
    UNION
    SELECT addr.user_id, user.name, addr.addr_name
    FROM user
    RIGHT JOIN addr
    ON user.user_id = addr.user_id
"""
pd.read_sql(QUERY, db)
```

```python
# SUB QUERY : SELECT, FROM, WJERE
# 전체 국가의 개수, 전체 도시의 개수를 1개의 row로 출력
QUERY = """select (select count(*) from country) as country
, (select count(*) from city) as city
"""
pd.read_sql(QUERY, db)
```

```python
# JOIN(4096*239=978,944) > Filtering
# 900만 이상이 되는 도시의 국가코드, 국가이름, 도시이름, 도시인구수 출력
QUERY = """
    SELECT city.countrycode, country.name, city.name, city.population
    FROM city
    JOIN country
    ON city.countrycode = country.code
    HAVING city.population >= 900*10000
""" 
pd.read_sql(QUERY, db)
```

```python
# FROM 절에 sub-query 쓰기 Filtering(6) > JOIN(6*239=1434)
# 900만 이상이 되는 도시의 국가코드, 국가이름, 도시이름, 도시인구수 출력
QUERY = """
    SELECT city.countrycode, country.name, city.name, city.population
    FROM (select * from city where population >= 900*10000) as city
    JOIN country
    ON city.countrycode = country.code
""" 
pd.read_sql(QUERY, db)
```

```python
QUERY = """
    SELECT countrycode
    FROM city
    WHERE population >= 900*10000
"""
pd.read_sql(QUERY, db)
```

```python
# WHERE 절에 sub-query 사용해보기
# 900만 이상이 되는 도시의 국가코드, 국가이름, 국가인구수 출력
QUERY = """
    select code, name, population
    from country
    where code in (
      select countrycode
      from city
      where population >= 900*10000
    )
""" 
pd.read_sql(QUERY, db)
```
