## Series

```python
from pandas import Series
# 다음 데이터에서 a와 n이 들어간 단어를 선택하라
s = Series(['apple', 'banana', 'melon', 'mango'])
s
```

```python
## 합집합구하는 경우
s[s.str.contains('a', 'n')]
```

```python
## 교집합
cond1 = s.str.contains('a')
cond2 = s.str.contains('n')
s[cond1 & cond2]
```

```python
# 합집합 반복문 사용해보기
for x in ['a', 'n']:
    cond = s.str.contains(x)
    cond += cond
s[cond]
## True + False -> True나옴 (1)
```

```python
# 교집합 반복문 사용해보기
# result = Series([True, True, True, True, True])
result = Series([True]*5)
for 변수 in ["a", "n"]:
    # result = result & s.str.contains(변수)
    result &= s.str.contains(변수)
print(result)
s[result] # 조건으로 슬라이싱
```
