# 클래스 - 5

상속 (inheritance)은 기존 클래스의 작동 방식을 변경함으로써 새로운 클래스를 만드는 매커니즘입니다.   
상속 또한 코드를 정리하고 재사용하는 방법입니다. 

## 상속

Parent라는 클래스는 sing() 메서드를 갖고 있습니다. 

````python
class 부모:
    def sing(self):
        print("가는세월~")

아빠 = 부모()
아빠.sing()
````

자식도 부모를 닮아 노래를 잘합니다. 

````python
class 자식:
    def sing(self):
        print("가는세월~")

사람 = 자식()
사람.sing()
````

위와 같은 경우 중복되는 코드가 존재합니다. 부모의 능력을 자식이 그대로 이어받으면 좋겠죠?

다음은 노래 잘 부르는 부모의 능력을 물려받은 자식 클래스를 정의합니다. 자식 클래스에는 `sing` 메서드가 정의돼 있지 않지만 부모 클래스의 메서드를 상속받아 사용할 수 있습니다. 

````python
class 자식(부모):
    pass

아들 = 자식()
아들.sing()
````

상속받지 못한 클래스는 당연하게도 `sing` 메서드를 사용할 수 없습니다. 

````python
class UnLuckyChild:
    pass

unluckyboy = UnLuckyChild()
unluckyboy.sing()
````

바람직한 운이 좋은 클래스를 만들어 봅시다. 이 친구는 부모의 능력을 이어 받아 노래도 부를 수있고 춤도 출 수 있습니다. 

````python
class LuckyChild(부모):
    def dance(self):
        print("shuffle dance")

luckyboy = LuckyChild()
luckyboy.sing()
luckyboy.dance()
````

````python
class 보병:
    def __init__(self):
        self.attack = 10
    
class 마린 :
    def __init__(self):
        self.attack = 10

class 파이어뱃:
    def __init__(self):
        self.attack = 10

a = 보병()
print(a.attack)
````

````python
# 상속으로 표현하기
class 보병: # 범용적기능이니까 부모로
    def __init__(self):
        self.attack = 10
    
class 마린(보병) : # 특정기능추가
    def steam(self):
        self.attack +=5

class 파이어뱃(보병):
    pass

a = 마린()
a.steam()
print(a.attack)
````

## 오버라이딩
부모클래스에 정의된 메서드를 재정의하는 것을 오버라이딩 (overriding)이라고 합니다.   
같은 이름으로 정의하면 생성한 객체 (`자식`)의 메서드가 실행됩니다. 

````python
class 부모:
    def 안녕(self):
        print("나는 부모야")
        
class 자식(부모):
    pass
````

부모에 정의된 메서드와 자식에 정의된 메서드가 다르다면?

````python
class 부모:
    def 안녕(self):
        print("나는 부모야")
        
class 자식(부모):
    def 안녕(self):
        print("나는 자식인데?")

자식().안녕()
````

자식 클래스의 메서드를 호출하고 부모 클래스의 메서드도 실행하고 싶다면 `super` 키워드를 사용합니다. `super`는 부모 클래스를 가리킵니다.

클래스의 호출한 결과는 어떻게 될까요?

````python
class 부모:
    def __init__(self, name):
        print("부모클래스")

class 자식(부모):        
    def __init__ (self, name):
        super().__init__(name)
        print("자식클래스")

객체 = 자식("종훈")
````

````python
class 부모:
    def __init__(self, name):
        print("부모클래스")

class 자식(부모):        
    def __init__ (self, name):
        print("자식클래스")
        super().__init__(name)        

객체 = 자식("종훈")
````
