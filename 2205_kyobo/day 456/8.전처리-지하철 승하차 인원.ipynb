## 지하철호선별 승하차 인원
data 폴더에는 `서울시 지하철호선별 역별 승하차 인원 정보.csv` 파일이 있습니다. `read_csv` 메서드를 사용하면 데이터프레임으로 값을 읽어올 수 있습니다.

Q. 승차총승객수와 하차총승객수는 몇명인가?

```python
df['승차총승객수'].sum()
```

Q. 승차총승객수가 가장 많은 "호선명"은 어디인가?

```python
df.groupby('호선명').agg(총승차=('승차총승객수','sum')).sort_values('총승차', ascending=False).head(5)
df.groupby('호선명').agg(총승차=('승차총승객수','sum')).idxmax()
```

Q. 승차총승객수가 가장 많은 역은 어디인가?

```python
df.groupby('역명').agg(총승차=('승차총승객수','sum')).sort_values('총승차', ascending=False)
df.groupby('역명').agg(총승차=('승차총승객수','sum')).idxmax()
```

Q. 승차총승객수와 하차총승객수의 합이 가장 많은 역과 가장 적은 역의 차이는?

```python
df['총승객수'] = df['승차총승객수'] + df['하차총승객수']
df.groupby('역명').agg(총승객수 = ('총승객수','sum')).sort_values('총승객수', ascending = False)
```

```python
a= df.groupby('역명').agg(총승객수 = ('총승객수','sum')).sort_values('총승객수', ascending = False).iloc[0]
b= df.groupby('역명').agg(총승객수 = ('총승객수','sum')).sort_values('총승객수', ascending = False).iloc[-1]
a-b
```

```python
a= df.groupby('역명').agg(총승객수 = ('총승객수','sum')).max()
b= df.groupby('역명').agg(총승객수 = ('총승객수','sum')).min()
a-b
```

Q. 승차총승객수와 하차총승객수의 합이 가장 많은 호선은?

```python
df.groupby('호선명').agg(호선별총승객수 = ('총승객수','sum')).sort_values('호선별총승객수', ascending = False)
df.groupby('호선명').agg(호선별총승객수 = ('총승객수','sum')).idxmax()
```

```python
df['일자'] = pd.to_datetime(df['사용일자'].astype(str))
df.head()
```

```python
# 요일 추가 (0:월 ~ 6:일)
df['요일'] = df['일자'].dt.weekday
df.head(2)
```

```python
df['평일주말'] = pd.cut(df['요일'], [0,5,6], labels=['평일','주말'])
df.head(2)
```

```python
df.groupby('평일주말').agg(승차수평균=('승차총승객수','mean'))
```
