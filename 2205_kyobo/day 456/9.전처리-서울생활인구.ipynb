Q. 컬럼의 이름을 짧게 수정하세요. 예:) 남15-19, 남20-24
- `columns` 속성에는 리스트 형태로 컬럼이름이 저장돼 있습니다. 
- 리스트의 값을 수정하면 컬럼 이름이 변경됩니다. 

```python
df.columns
```

```python
list_col = []
for column in df.columns:
    if "세부터" in column:
        a = column.replace("자","").replace("세부터","-").replace("세생활인구수","")
        list_col.append(a)
    elif "세이상" in column:
        a = column.replace("자","").replace("세이상생활인구수","-99")
        list_col.append(a)
    else:
        list_col.append(column)
df.columns = list_col
df.head(2)
```
