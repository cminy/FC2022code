### 리스트 인덱싱과 슬라이싱

데이터는 저장된 순서대로 0부터 시작하는 인덱스가 차례로 부여됩니다.  
슬라이싱에 사용되는 인덱스는 데이터와 데이터 사이에 위치합니다.  
문자열 인덱싱/슬라이싱과 원리가 같습니다.  

<img src="https://i.ibb.co/tHTjCzM/2-1.png" style="float:left" />

````python
## 리스트는 변수와 다르게 인덱스도 메모리를 차지함
# [0 1 2 삼성전자 LG전자 네이버] 요런식
````

````python
interest = ["삼성전자", "LG전자", "카카오"]
````

인덱스가 있으니 인덱싱과 슬라이싱을 할 수 있겠죠?  
문자열을 슬라이싱하면 문자열! 리스트를 슬라이싱하면 리스트!

````python
interest = ["삼성전자", "LG전자", "카카오"]
# 슬라이싱 : 값 사이사이에 번호가 들어감
## 삼성전자 하나 슬라이싱:
interest[0:1] # 슬라이싱 -> 원래꺼 타입대로. 리스트면 슬라이싱 해도 리스트로.
````

````python
# 인덱싱
interest[1], interest[1][0:2] # 인덱싱은 문자열로 반환됨.
````

인덱스를 벗어나면?

````python
interest[4]
````

문자열과 달리 리스트의 값은 수정이 가능합니다.  
아래는 리스트의 1번 인덱스의 값을 변경하는 코드입니다.

###### 연습문제 - 3
리스트에는 집주소로 정렬된 거주인원의 이름이 저장돼 있습니다.   
`| 아름 | 종훈 | 현호 | 시후 | 아라 | 정현 |`  

----

하나의 이름을 입력받아 인접한 두 집을 출력하세요.  
`name = input("이름:")`

----

```python
# 출력 예시
>> 이름: 현호
주변에는 종훈 시후가 살고 있습니다.
```

````python
거주순서 = ["아름", "종훈", "현호", "시후", "아라", "정현"]
name = input("이름:")
idx_s = 거주순서.index(name)
if idx_s == 0:
  print(f"주변에는 {거주순서[idx_s+1]}가 살고 있습니다.")  
elif idx_s == len(거주순서)-1:
  print(f"주변에는 {거주순서[idx_s-1]}가 살고 있습니다.") 
else:
  print(f"주변에는 {거주순서[idx_s-1]} 와 {거주순서[idx_s+1]}가 살고 있습니다.")
````
