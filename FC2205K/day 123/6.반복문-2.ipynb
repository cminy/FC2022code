## 반복문 - 응용형
반복문의 자료구조 자리에 인덱스를 넣어주고 인덱싱으로 원하는 값을 출력할 수 있습니다. 

<img src="https://i.ibb.co/xYbWFpK/for-1.png" width="450" style="float:left" />

`i`는 변수이기 때문에 이름을 변경해도 같은 결과를 출력합니다. 가독성에 좋은 변수를 사용하는게 좋겠죠?

````python
stocks = ["Naver", "Samsung", "SK Hynix"]
for 인덱스 in [0, 1, 2] :
    print(stocks[인덱스])
````

인덱스를 직접 입력해도 좋지만 range 함수를 사용하면 많은 숫자의 인덱스도 쉽게 만들 수 있습니다. 

````python
for 인덱스 in range(0,3):
    print(인덱스)
````

슬라이싱과 유사하게 시작 인덱스, 증가폭을 지정해줄 수 있습니다. 다음은 1부터 2까지의 인덱스를 출력합니다

````python
for 인덱스 in range(1,3,1):
    print(인덱스)
````

세 개의 파라미터는 증감폭을 지정합니다. 

len 함수까지 사용하면 코드의 재사용성을 높일 수 있습니다.  
데이터가 추가되더라도 반복문의 수정없이 모든 값이 출력됩니다. 

````python
stocks = ["Naver", "Samsung", "SK Hynix"]

for i in range(0,len(stocks)):
  print(stocks[i])
````

빈번하게 사용되는 형태이므로 `for 인덱스 in range(len(자료구조)):` 코드를 외우는 것도 좋습니다. 인덱스를 사용하는 for문은 어려워 보이지만 자료구조에 저장된 여러 데이터를 한 번에 가져올 수 있어서 사용합니다. 다음 예제를 살펴 봅시다.

````python
리스트 = ["가", "나", "다", "라"]

for i in range(len(리스트)):
  print(리스트[i])
````

인덱스를 조정해서 한 번에 두 개의 데이터를 출력합니다.

````python
for i in range(len(리스트)-1): # = [0,1,2] = range(3)
  print(리스트[i]+리스트[i+1])
````

````python
리스트 = ["가", "나", "다", "라", "마", "바", "사"]
for i in range(1, len(리스트)): 
  print(리스트[i-1]+리스트[i])
````

Q. `For 변수 in range` 문법을 사용해서 아래와 같이 화면에 출력하라.  
```
0 1000
1 400
2 2000
3 2400
```

````python
price = [1000, 400, 2000, 2400]

for i in range(len(price)):
  print(i, price[i])
````

````python
price = [1000, 400, 2000, 2400]
for i, num in enumerate(price):
  print(i, num)
````

Q. 3일 이동 평균을 출력하라.

````python
price = [100, 200, 300, 400, 500, 600, 700]
pr_mean = []

for i in range(len(price)-2):
  print(sum(price[i:i+3])/3)

print('------'*2)

for i in range(1,len(price)-1):
  print(sum(price[i-1:i+2])/3)

print('------'*2)

for i in range(0,len(price)):
  if i < 2 :
    pr_mean.append(sum(price[0:3])/3)
  else :
    pr_mean.append(sum(price[i-2:i+1])/3)
len(pr_mean), pr_mean
````

````python
# 리스트 연산 이용하기
r = []

for i in range(1,len(price)-1):
  r.append(sum(price[i-1:i+2])/3)
print(r)

print(r[:1]*2+r)

r = r[:1]*(len(price)-len(r)) + r
print(r)
````

````python
price = [10, 100, 30, 200, 400, 60, 500,  1000, 2000]
result = []
tab = 4 # 4일 이동평균
for i in range(len(price)-tab+1):
  result.append(sum(price[i:i+tab])/tab)
print(result)
# y = result[:1]*(len(price)-len(result)) + result
y = result[:1]*(tab -1) + result
print(y)
````

````python
import matplotlib.pyplot as plt
plt.plot(range(len(price)), price)
plt.plot(range(len(price)),y)
````

### enumerate
인덱스와 값을 한 번에 전달해주는 함수

````python
자료구조 = ["가", "나", "다"]
for idx, val in enumerate(자료구조):
    print(idx, val)
````
