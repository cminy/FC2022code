# 파일 다루기 (쓰기)

메모리의 데이터는 컴퓨터

```python
f = open("data.txt", "w") # w 쓰기모드
```

```python
f.write("안녕하세요.") # 결과 6 > 여섯글자썼다구
```

```python
f.close() # close 해야 저장되지
```

```python
# 항상 open - close 귀찮으니까 짧게 바꿔보자
with open("data.txt", "w") as f:
    f.write("반가워ㅋㅋㅋ")
```

```python
# 이어쓰기 a
with open("data.txt", "a") as f:
    f.write("나야나")
```

```python
with open("data.txt", "a") as f:
    f.write("""
    안녕 나는 최민영 요요
    예예
    안녕하세요.
    반갑습니다.
    """)
```
