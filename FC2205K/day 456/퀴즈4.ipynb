# 문제-3
다음의 데이터프레임에서 데이터 타입을 변경하세요.
- `현재가` 컬럼은 `int64`로 `등락률` 컬럼은 `float64`로 변경

```python
from pandas import DataFrame

data = [
    ["037730", "3R", "1,510", "7.36"],
    ["036360", "3SOFT", "1,790", "1.65"],
    ["005760", "ACTS", "1,185", "1.28"],
]
columns = ["종목코드", "종목명", "현재가", "등락률"]
df = DataFrame(data=data, columns=columns)
df
```

```python
df.info()
```

```python
df['현재가'] = df['현재가'].str.replace(',','') # 문자열 replace
df
# 다른 방법들
# 1) Series의 replace : df['현재가'].replace(",", "", regex=True)
# 2) map 함수 사용 : 
# def f(x):
#     return x.replace(".". "")
# df['현재가'].map(f)
```

```python
how = {   
    "현재가"  : 'int64',
    "등락률" : 'float64'
}

df = df.astype(how)
```

```python
df.info()
```

# 문제-4

```python
from pykrx import stock
df = stock.get_market_ohlcv("20190101", "20220528", "005930")
df.head()
```

```python
df['연도'] = df.index.year
df.groupby('연도')['종가'].mean()
```

```python
# agg 써보기
df.groupby('연도').agg(mean = ('종가','mean'))
```

```python
# agg에 딕셔너리 넣어서 써보기
how = {
    '종가': np.mean
}
df.groupby('연도').agg(how)
```

- 출력되는 데이터프레임에는 두 개의 컬럼(`평균`, `거래량`)이 있어야 합니다.

```python
df.index.month
```

```python
df.index.year
```

```python
import numpy as np
df['월'] = df.index.month
df.groupby(['연도','월']).agg(평균=('종가','mean'),거래량=('종가','sum'))
```
