# 반복문
반복문은 자료구조의 데이터 수만큼 코드를 반복실행 합니다.  
자료구조에 있는 "데이터를 하나씩 바인딩하고 들여쓰기된 코드를 실행" 과정을 반복하는 겁니다. 
  
## 반복문 - 기본형
`for 변수 in 자료구조 : ` 형태로 코드를 작성합니다. 코드 끝에 콜론 `:` 있음에 주의하세요. 분기문과 같이 콜론 다음에는 들여쓰기와 함께 영향을 받는 코드를 작성합니다. 

<img src="https://i.ibb.co/gR952tp/for-0.png" width="300" style="float:left" />

###### 반복문을 사용하지 않고 한 줄에 하나의 값 출력하기

````python
리스트 = ["a","b","c","d"]
````

````python
print(리스트[0])
print(리스트[1])
print(리스트[2])
print(리스트[3])
````

###### 반복문으로 코드를 정리하기

````python
for _ in 리스트:
  print(_)
````

###### 반복문 사용 예)
`y = x ** 2` 그래프를 그려보자

````python
x = [-2, -1, 0, 1, 2]
y = []

for xx in x:
  y.append(xx**2)
print(y)
````

````python
# 시각화
import seaborn as sns
import matplotlib.pyplot as plt

plt.plot(x,y)
````

###### 연습문제 - 4
`xls` 파일만을 출력하라

````python
리스트 = ["ab.xls", "a.hwp", "b.exe", "bbbbb.xls", "abcd.axls"]
x, y, z, zz = [],[],[], []
for 변수 in 리스트:
  # 방법1
  if 변수[-4:] == ".xls":
      x.append(변수)
  # 방법2
  if 변수.split('.')[1] == "xls":
      y.append(변수)
  # 방법3
  if ".xls" in 변수:
      z.append(변수)
  # 방법4
  if 변수.endswith('.xls'):
      zz.append(변수)
print(x,y,z,zz, sep="\n---\n")
````

###### 연습문제 - 6
반복문을 사용해서 변수에 저장된 종목을 화면에 출력하라.  
```
보유종목:Naver 
--------------
보유종목:Samsung
--------------
보유종목:SK Hynix
--------------
```

###### 연습문제 - 7
반복문을 사용해서 변수에 저장된 종목을 화면에 출력하라.   
```
1999년 성적: 수 
2000년 성적: 양 
2001년 성적: 미 
2002년 성적: 수 
```

````python
my_grade = ["1999-수", "2000-양", "2001-미", "2002-수"]

for year_grade in my_grade:
  print(f"{year_grade[:4]}년 성적: {year_grade[-1]}")
````
